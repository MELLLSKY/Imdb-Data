```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('./NLPlabeledData.tsv' , delimiter = '\t' , quoting = 3)
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
# Download stopwords from NLTK
nltk.download("stopwords")
```

```python
from bs4 import BeautifulSoup
# Function to clean HTML tags using BeautifulSoup
def clean_html_with_bs4(html_text):
    # clean html tags
    soup = BeautifulSoup(html_text, "html.parser")
    return soup.get_text()

# Example of cleaning HTML from a review
with_html = df.review[0]
cleaned_text = clean_html_with_bs4(with_html)
print(cleaned_text)
```

```python
import re
#clean stopwords
cleaned_text = re.sub('[^a-zA-Z]', ' ', cleaned_text)

#clean uppercase
cleaned_text = cleaned_text.lower()

#split
cleaned_text = cleaned_text.split()

print(cleaned_text)
```

```python
len(cleaned_text)
```

```python
# Remove stopwords
swords = set(stopwords.words('english'))
cleaned_text = [w for w in cleaned_text if w not in swords]
print(cleaned_text)
```

```python
len(cleaned_text)
```

```python
# Function to perform full preprocessing on a review
def opr(review):
  review = BeautifulSoup(review).get_text()
  review = re.sub('[^a-zA-Z]', '' , review)
  review.lower()
  review = review.split()
  swords = set(stopwords.words('english'))
  review = [w for w in review if w not in swords]
  return(''.join(review))
```

```python
train_x_tum = []
for i in range(len(df['review'])):
  if (i+1)%1000==0:
    print('step', i+1)
  train_x_tum.append(opr(df['review'][i]))
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Prepare the features (X) and labels (y)
x = train_x_tum
y = np.array(df['sentiment'])
```

```python
# Split the data into training and testing sets
train_x, test_x, train_y, test_y =train_test_split(x, y, test_size=0.15, random_state=33)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Vectorize the text data (convert text to numerical features)
vectorizer = CountVectorizer(max_features = 5000)
train_x1 = vectorizer.fit_transform(train_x)
train_x1
```

```python
# Convert the sparse matrix to a dense array
train_x1 = train_x1.toarray()
train_y1 = train_y
train_x1.shape, train_y1.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize a Random Forest Classifier
rf = RandomForestClassifier(n_estimators= 100)
```

```python
# Train the model on the training data
rf.fit(train_x1, train_y1)
```

```python
# Transform the test data using the same vectorizer
test_x1 = vectorizer.transform(test_x)
test_x1
```

```python
test_x1 = test_x1.toarray()
test_x1.shape
```

```python
# Calculate the accuracy of the model using the ROC-AUC score
from sklearn.metrics import roc_auc_score
test_predict = rf.predict(test_x1)
accuracy = roc_auc_score(test_y, test_predict)
accuracy
```
